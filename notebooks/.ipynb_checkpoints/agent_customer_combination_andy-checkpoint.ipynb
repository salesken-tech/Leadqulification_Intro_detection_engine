```python
import pandas as pd
import time
```

```python
df = pd.read_excel("/home/andy/Desktop/byjoose_snippets.xlsx")
```

```python
df.shape
```

```python
start = time.time()
cached_snippets=[]
for i in range(len(df)):
    if len(cached_snippets) ==0:
        print("here")
        cached_snippets.append({"orignal_ids":[df["id"][i]],"speaker":df["speaker"][i],"text":df["text"][i],"from_time":df["from_time"][i],"to_time":df["to_time"][i],"task_id":df["task_id"][i]})
        print(cached_snippets,i)
    else:
        if df["speaker"][i] == cached_snippets[-1]['speaker']:
            cached_snippets[-1]["orignal_ids"].append(df["id"][i])
            cached_snippets[-1]["text"]+=". "+df["text"][i]
            cached_snippets[-1]["to_time"]=df["to_time"][i]
            cached_snippets[-1]["task_id"]=df["task_id"][i]
        else:
            cached_snippets.append({"orignal_ids":[df["id"][i]],"speaker":df["speaker"][i],"text":df["text"][i],"from_time":df["from_time"][i],"to_time":df["to_time"][i],"task_id":df["task_id"][i]})
        
print("Time for making new snippet for 1822 snippets ={}".format(time.time()-start))
    
```

```python
new_snippets_df=pd.DataFrame(cached_snippets)
```
